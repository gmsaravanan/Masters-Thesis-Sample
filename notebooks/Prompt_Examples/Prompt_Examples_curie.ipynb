## Fact Probing



```python
prompt = (
    "The capital of India is: "
)

response = openai.Completion.create(model=model, prompt=prompt, temperature=0.1, max_tokens=7)
print(response.choices[0].text)
```

## Text Classification



```python
prompt = (
    "Predict sentiment of text: \n"
    "text: Really loved the movie. sentiment: "
)

response = openai.Completion.create(model=model, prompt=prompt, temperature=0.1, max_tokens=10)
print(response.choices[0].text)
```

```python
prompt = (
    "Predict sentiment for text: \n"
    "text: Really loved the movie. sentiment: POSITIVE. \n"
    "text: The movie was complete waste of time. sentiment: NEGATIVE. \n"
    "text: Awesome, I want to go watch it again. sentiment: "
)

response = openai.Completion.create(model=model, prompt=prompt, temperature=0.1, max_tokens=7)
print(response.choices[0].text)
```

## Question Answering



```python
prompt = (
    "Generate answer to the question given the passage: \n"
    "passage: Independence Day is celebrated annually on 15 August as a public holiday in India commemorating the nation's independence from the United Kingdom on 15 August 1947. "
    "question: When is the Independence day of India? "
)

response = openai.Completion.create(model=model, prompt=prompt, temperature=0.1, max_tokens=15)
print(response.choices[0].text)
```
